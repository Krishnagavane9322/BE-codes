```python
#Use MNIST Fashion Dataset and create a classifier to classify fashion clothing into categories
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
model = keras.Sequential([
    keras.Input(shape=(28, 28)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```

```python
loss, acc = model.evaluate(x_test, y_test)
print("\nTest Accuracy:", acc)
```

```python
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]
```

```python
sample = x_test[0]
prediction = model.predict(sample.reshape(1, 28, 28))

predicted_label = class_names[prediction.argmax()]
actual_label = class_names[y_test[0]]

print("Prediction:", predicted_label)
print("Actual:", actual_label)
```

```python
plt.imshow(sample, cmap='gray')
plt.title(f"Predicted: {predicted_label}")
plt.show()
```
